## **Introducción**

Los sistemas de transporte representan para las grandes ciudades una herramienta importante para mejorar la movilidad urbana, reducir la congestión vehicular y promover alternativas de movilidad más sustentables. Resulta necesario impulsar soluciones de transporte que permitan a la población desplazarse de manera eficiente, accesible y amigable con el medio ambiente, especialmente en ciudades con alta densidad poblacional y grandes retos de movilidad.

En la CDMX, el sistema ECOBICI se ha consolidado como una alternativa de transporte que permite a miles de personas realizar trayectos diarios de manera rápida y eficiente, reduciendo los tiempos que el tráfico automovilistico representa y las incomodidades de metros y camiones saturados.

La importancia de este análisis radica en que los datos pueden convertirse en información útil para la toma de decisiones.  **2. LIMPIEZA DE DATOS**

**Valores faltantes**



De acuerdo con la información recabada hasta el momento, es importante tomar en consideración los siguientes puntos con respecto a los valores faltantes:

*   existe 1 valor nulo en Genero_Usuario.
*   existen 79 valores nulos en Edad_Usuario.

La presencia de valores faltantes únicamente contempla las variables Genero_Usuario y Edad_Usuario. Debido a que representan una cantidad mínima respecto al total de registros de toda la BD, se evaluó la estrategia adecuada para su tratamiento sin afectar significativamente el volumen de información disponible.

Durante la revisión de los valores faltantes en la variable Edad_Usuario, se identificaron 79 registros sin información de edad, lo que representa una proporción mínima respecto al total de observaciones del conjunto de datos. Debido a que estos registros contienen información relevante sobre los viajes realizados, como estaciones de origen y destino, horarios y bicicletas utilizadas, se decidió conservarlos en lugar de eliminarlos, evitando perder información útil para análisis posteriores relacionados con los patrones de movilidad dentro del sistema ECOBICI.

Para completar los valores faltantes de edad se aplicó una estrategia de imputación utilizando la mediana de la variable, en lugar de la media. Esta decisión se tomó debido a que la edad presenta un rango amplio de valores, incluyendo posibles edades extremas, y la mediana es una medida estadística menos sensible a estos valores atípicos. De esta manera, se mantiene una representación más adecuada del usuario promedio sin alterar significativamente la distribución original de los datos.




*  **Edades faltantes**

**Rellenar edades faltantes con la mediana**

```python
df_clean = df.copy() #Hacemos una copia por seguridad para guardar la BD limpia

mediana_edad = df_clean["Edad_Usuario"].median() # encontramos la mediana en la columna "Edad_Usuario" del df_clean y lo guardamos en la variable mediana_edad. Pandas revisa todos los valores de la columna Edad_Usuario, los ordena y obtiene el valor central.
print(mediana_edad)

df_clean["Edad_Usuario"] = df["Edad_Usuario"].fillna(mediana_edad) #En la columna Edad_Usuario, donde se encuentre valores vacíos, coloca la mediana. fillna() no modifica la columna original automáticamente, por eso es necesario ejecturar df["Edad_Usuario"] = ... para guardar el resultado nuevamente en la columna.
```

**Verificar que ya no existan valores nulos en Edad_Usuario**

```python
print(f"Edades nulas en df_clean: {df_clean['Edad_Usuario'].isnull().sum()}") #En la columna "Edad_Usuario" del df_clean ejecuta isnull().sum() para sumar los valores nulos

print("Edades nulas en df:", df['Edad_Usuario'].isnull().sum()) #A diferencia del df_clean, el original df continúa con valores nulos
```

*  **Género faltante**

De 1,525,302 registros, solamente 1 registro no tiene información de género. Debido a que representa un porcentaje insignificante respecto al total de registros y que no existe una forma objetiva de determinar la categoría correcta sin introducir información artificial, se decidió eliminar éste. Esta acción permitió mantener únicamente datos completos para los análisis relacionados con características demográficas de los usuarios.

```python
df_clean = df_clean.dropna(subset=["Genero_Usuario"]) #Esto elimina únicamente las filas donde Genero_Usuario está vacío, sin afectar otras columnas.

#dropna(): elimina las filas o columnas que contienen valores nulos (NaN) en un DataFrame.
#subset: indica las columnas específicas que dropna() debe revisar para encontrar valores nulos.
```

* **Edades atípicas**

Durante el análisis exploratorio de la variable Edad_Usuario, se identificaron registros con edades elevadas, incluyendo valores superiores a los 90 años. Aunque estas edades son biológicamente posibles, representan casos poco frecuentes dentro del contexto de un sistema de bicicletas compartidas, por lo que se consideraron valores potencialmente atípicos que requieren una revisión adicional.

Debido a que la base de datos corresponde a registros de viajes y no a usuarios únicos, la frecuencia observada representa la cantidad de viajes realizados por personas con determinada edad registrada, y no necesariamente el número de usuarios. Por esta razón, se decidió conservar estos registros para evitar la pérdida de información, pero se creó la variable auxiliar Edad_Atipica, la cual permite identificar y analizar por separado aquellos viajes asociados a edades iguales o superiores a 90 años. Esta clasificación facilita evaluar el impacto de estos valores en los análisis posteriores sin asumir automáticamente que se trata de errores de captura.

# 4. Manipulación de datos


Durante la etapa de preparación de datos, se realizó la conversión de las columnas Hora_Retiro y Hora_Arribo de texto (object) a un formato temporal reconocido por Pandas. Adicionalmente, se propone la creación de las columnas derivadas Inicio_Viaje, Fin_Viaje, Duracion_Minutos y Dia_Semana, sin eliminar las variables originales de fecha y hora. Esto facilita el cálculo de la duración de los recorridos y permite realizar análisis más avanzados sobre tiempos de traslado, eficiencia del servicio y patrones de movilidad. Las columnas originales se conservan para mantener la información de origen y las nuevas variables funcionan como indicadores calculados para el análisis.

```python
df_clean = df_clean.rename(columns={
    "Ciclo_Estacion_Retiro": "Estacion_Retiro",
    "Ciclo_EstacionArribo": "Estacion_Arribo"
})
```

```python
# Contamos el número de retiros y arribos en cada estación
conteo_retiros = df_clean["Estacion_Retiro"].value_counts()
conteo_arribos = df_clean["Estacion_Arribo"].value_counts()

# Creamos un DataFrame para comparar ambos conteos
df_balances = pd.DataFrame({
    "Retiros": conteo_retiros,
    "Arribos": conteo_arribos
})

# Si existe un Nan lo llenamos con 0
df_balances = df_balances.fillna(0)

# Calculamos la diferencia (Si el resultado es positivo, existen más devoluciones que retiros; si es negativo, existen más retiros que devoluciones.)
df_balances["Balance_Neto"] = df_balances["Arribos"] - df_balances["Retiros"]

# Calculamos el balance. El balance perfecto sería 1, ya que implicaría que por cada bici retirada, hay una que se devuelve a la misma estación (entre más se aleje el resultado de 1, mayor diferencia entre retiros y devoluciones existe)
df_balances["Proporcion_Arribo_Retiro"] = df_balances["Arribos"] / df_balances["Retiros"]

# Creamos una tabla de catálogo que asocie cada ID de estación con su dirección única (eliminando los duplicados)
catalogo_estaciones = df_clean[['Estacion_Retiro', 'direccion_origen']].drop_duplicates()

# Juntamos la tabla de balances con el catálogo usando el ID de la estación
df_balances = pd.merge(df_balances, catalogo_estaciones, left_on=df_balances.index, right_on='Estacion_Retiro')

# Colocamos el ID de la estación como el índice principal de nuestra tabla de balances
df_balances = df_balances.set_index('Estacion_Retiro')

# Filtramos estaciones con más de 500 retiros para prescindir de las estaciones menos utilizadas
estaciones_activas = df_balances[df_balances["Retiros"] > 500]
# Definimos las columnas que queremos mostrar en el reporte final (ahora usando 'direccion_origen')
columnas_reporte = ["direccion_origen", "Retiros", "Arribos", "Balance_Neto", "Proporcion_Arribo_Retiro"]

print("Las 5 estaciones con mayor excedente de devoluciones (constantemente llenas):")
# Mostramos las columnas clave ordenadas de mayor a menor balance
display(estaciones_activas.sort_values(by="Balance_Neto", ascending=False)[columnas_reporte].head(5))

print("\nLas 5 estaciones con mayor excedente de retiros (constantemente vacías):")
# Ordenadas de menor a mayor balance (los más negativos)
display(estaciones_activas.sort_values(by="Balance_Neto", ascending=True)[columnas_reporte].head(5))
```

# **Análisis del tiempo de los trayectos**

```python
# 1. Convertimos primero las columnas de hora a un formato temporal reconocido por Pandas
df_clean["Hora_Retiro"] = pd.to_datetime(df_clean["Hora_Retiro"], errors='coerce')
df_clean["Hora_Arribo"] = pd.to_datetime(df_clean["Hora_Arribo"], errors='coerce')

# Combinamos la fecha con la hora extraída correctamente usando .dt
df_clean["Inicio_Viaje"] = (
    df_clean["Fecha_Retiro"].astype(str)
    + " "
    + df_clean["Hora_Retiro"].dt.strftime("%H:%M:%S")
)
df_clean["Inicio_Viaje"] = pd.to_datetime(df_clean["Inicio_Viaje"])

# ------------------------------------------------------------------------------
df_clean["Fin_Viaje"] = (
    df_clean["Fecha_Arribo"].astype(str)
    + " "
    + df_clean["Hora_Arribo"].dt.strftime("%H:%M:%S")
)
df_clean["Fin_Viaje"] = pd.to_datetime(df_clean["Fin_Viaje"])
```

```python
# Crear columna de duración

df_clean["Duracion_Minutos"] = (
    df_clean["Fin_Viaje"] - df_clean["Inicio_Viaje"]
).dt.total_seconds() / 60
```

```python
df_clean["Dia_Semana"] = df_clean["Inicio_Viaje"].dt.day_name() #Crear columna Dia_Semana
```

## **Análisis de Patrones Temporales y Ciclos de Demanda**

**Patrones de Demanda Temporal y Rutinas de Movilidad**

**Distribución de la Demanda por Día de la Semana**
Al ordenar los viajes de mayor a menor volumen, se observa que la actividad del sistema se concentra de manera masiva de lunes a viernes, registrando sus niveles más altos a mitad de la semana (miércoles y jueves). Esta desaceleración dominical refleja el cese de las actividades de oficina y el cambio hacia una dinámica de movilidad puramente local o de esparcimiento.

A nivel general, el volumen de viajes se concentra de forma crítica en dos franjas horarias que coinciden con el inicio y término de las jornadas habituales:

Activación de Estaciones de Baja Movilidad: Para las estaciones donde las bicicletas permanecen fijas por mucho tiempo, se deben evaluar estrategias de reactivación, tales como su reubicación hacia puntos con mayor flujo comercial o el rediseño de la infraestructura ciclista a su alrededor para hacerlas más accesibles.

Infraestructura con Perspectiva de Género
Construcción de Calles Seguras: Para reducir la brecha actual (donde los hombres realizan el 67.5% de los viajes frente al 27.4% de las mujeres), es urgente diseñar cruces peatonales y ciclistas mejor iluminados, así como ciclovías completamente separadas del flujo vehicular en avenidas principales. Estudios de organismos como el ITDP y el BID respaldan que una infraestructura protegida disminuye la percepción de inseguridad vial, siendo el motor principal para que más mujeres adopten la bicicleta en su día a día.

Inclusión de Nuevos Sectores de la Población
Incentivos Económicos y Sociales: Como la participación de menores de 18 años y de personas de la tercera edad es muy baja, el sistema debería implementar tarifas preferenciales (como descuentos estudiantiles o vinculación directa con la tarjeta de INAPAM).

Talleres Comunitarios: Complementar los incentivos económicos con talleres gratuitos de ciclismo urbano comunitario facilitará que estos sectores pierdan el miedo a rodar en la ciudad y comiencen a utilizar ECOBICI para sus viajes cortos, escolares o recreativos durante los fines de semana, aprovechando que en esos días la curva de demanda es más suave y relajada.
